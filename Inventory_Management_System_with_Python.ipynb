```python

def display_inventory(inventory):
    print("Inventory list:")
    if not inventory:
        print("No items in inventory.")
    else:
        for student in inventory.values():
            print(f"Name: {student['name']}, age: {student['age']}, GPA: {student['gpa']}, hobby: {student['hobby']}")


def add_to_inventory(inventory):
    name = input("Enter student name: ")
    age = int(input("Enter student age: "))
    gpa = float(input("Enter student GPA: "))
    hobby = input("Enter student hobby: ")
    inventory[name] = {'name': name, 'age': age, 'gpa': gpa, 'hobby': hobby}
    print("Student added to inventory successfully!")


def search_inventory(inventory):
    name = input("Enter student name to search: ")
    student = inventory.get(name)
    if student:
        print(f"Name: {student['name']}, age: {student['age']}, GPA: {student['gpa']}, hobby: {student['hobby']}")
    else:
        print("Student not found in inventory.")


def delete_inventory(inventory):
    name = input("Enter student name to delete: ")
    if name in inventory:
        del inventory[name]
        print("Student deleted from inventory successfully!")
    else:
        print("Student not found in inventory.")


inventory = {}

while True:
    print("\nOptions:")
    print("1. List all Inventory")
    print("2. Add to Inventory")
    print("3. Search Inventory")
    print("4. Delete Inventory")
    print("5. Exit")

    option = input("\nSelect an option: ")

    if option == "1":
        display_inventory(inventory)
    elif option == "2":
        add_to_inventory(inventory)
    elif option == "3":
        search_inventory(inventory)
    elif option == "4":
        delete_inventory(inventory)
    elif option == "5":
        break
    else:
        print("Invalid option. Please try again.")
```
